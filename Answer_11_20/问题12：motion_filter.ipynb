```python
import cv2
import numpy as np
```

```python
def motion_filter(img,kernel): 
    half_ksize = int(kernel.shape[0]/2)
    img_new=np.copy(img)
    for c in range(3):
        for i in range(1,img.shape[0]-1):#去掉边缘
            for j in range(1,img.shape[1]-1):
                img_new[i,j,c]=np.sum(kernel*img[i-half_ksize:i+half_ksize+1,j-half_ksize:j+half_ksize+1,c])
    return img_new.astype(np.uint8)
```

```python
kernel = np.array([[1/3,0,0],[0,1/3,0],[0,0,1/3]])
```

```python
img = cv2.imread(r'D:\Program Files (x86)\ImageProcessing100Wen-master\ImageProcessing100Wen-master\Question_01_10\imori.jpg')
cv2.imshow('1',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
output = motion_filter(img,kernel)#核大小为8*8
cv2.imshow('2',output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
