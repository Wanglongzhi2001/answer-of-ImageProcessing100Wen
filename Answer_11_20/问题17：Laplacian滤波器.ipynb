```python
import cv2
import numpy as np
```

```python
def BGR2GRAY(img):
    b,g,r = cv2.split(img)
    gray = (0.2126 * r + 0.7152 * g + 0.0722 * b).astype('uint8') #一定要变成uint8哦
    return gray
```

```python
def Laplace_filter(img,kernel): #拉普拉斯锐化操作
    img_new=np.copy(img)
    half_kshape = kernel.shape[0]//2
    r,w=img.shape
    for i in range(half_kshape,r-half_kshape):
        for j in range(half_kshape,w-half_kshape):
            R=np.sum(kernel*img[i-half_kshape:i+half_kshape+1,j-half_kshape:j+half_kshape+1])
            img_new[i,j]=img[i,j]-R
    img_new = np.clip(img_new,0,255)
    return img_new
```

```python
kernel=np.array([[1,1,1],[1,-8,1],[1,1,1]]) #拉普拉斯模板
img = cv2.imread(r'D:\Program Files (x86)\ImageProcessing100Wen-master\ImageProcessing100Wen-master\Question_01_10\imori.jpg')
img_gray = BGR2GRAY(img)
```

```python
output = Laplace_filter(img_gray,kernel)
cv2.imshow('2',output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
