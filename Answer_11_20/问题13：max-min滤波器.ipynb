```python
import cv2
import numpy as np
```

```python
def BGR2GRAY(img):
    b,g,r = cv2.split(img)
    gray = (.2126 * r + 0.7152 * g + 0.0722 * b).astype('uint8') #一定要变成uint8哦
    return gray
```

```python
def max_min_filter(img_gray,k_size=3):
    output = img_gray.copy()
    half_ksize = int(k_size/2)
    for i in range(half_ksize,img_gray.shape[0]-half_ksize):
        for j in range(half_ksize,img_gray.shape[1]-half_ksize):
            max_v = np.max(img_gray[i-half_ksize:i+half_ksize+1,j-half_ksize:j+half_ksize+1])
            min_v = np.min(img_gray[i-half_ksize:i+half_ksize+1,j-half_ksize:j+half_ksize+1])
            output[i,j] = max_v - min_v
    return output
```
