```python
import cv2
import numpy as np
```

```python
def Gaussian_filter(img,kernel): #delta=3
    img = np.array(img, dtype='int')  # 转换成int型防止像素加减溢出
    img_new=np.copy(img)
    for c in range(3):
        for i in range(1,img.shape[0]-1):#去掉边缘
            for j in range(1,img.shape[1]-1):
                img_new[i,j,c]=np.sum(kernel*img[i-1:i+2,j-1:j+2,c])
    return img_new.astype(np.uint8)
```

```python
kernel = np.array([[1,2,1],[2,4,2],[1,2,1]])/16
```

```python
img = cv2.imread(r'D:\Program Files (x86)\ImageProcessing100Wen-master\ImageProcessing100Wen-master\Question_01_10\imori_noise.jpg')
cv2.imshow('1',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
output = Gaussian_filter(img,kernel)#核大小为8*8
cv2.imshow('2',output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
