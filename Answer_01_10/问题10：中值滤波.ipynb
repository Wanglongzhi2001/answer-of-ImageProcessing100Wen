```python
import cv2
import numpy as np
```

```python

def median_filter(img,kernel_size):
    img_new=np.copy(img)
    half_ksize = int(kernel_size/2)
    for c in range(3):
        for i in range(1,img.shape[0]-1):#去掉边缘
            for j in range(1,img.shape[1]-1):
                img_new[i,j,c]=np.median(img[i-half_ksize:i+half_ksize+1,j-half_ksize:j+half_ksize+1,c])
    return img_new.astype(np.uint8)
```
